```python

    #Importamos la libería keyword
import keyword

    #Entrada del usuario
cadena_11 = (input("Introduzca una cadena de texto: "))

#Se crean las siguientes cadenas a usar:
vocales = 'AEIOU'                #Vocales mayúsculas
vocales_tilde = 'ÁÉÍÓÚ'          #Vocales mayúsculas con tilde
vocales_tilde_minus = 'áéíóú'    #Vocales minúsculas con tilde
cad_space = ' '                  #Cadena espacio
reserv = (keyword.kwlist)        #Se importa la función kwlist, que contiene 
                                 #las palabras reservadas

          #Se convierte a lista la cadena ingresada por el usuario
lista_11 = cadena_11.split(' ')


#Se definen los contadores:
cont_tildes = 0      #Cuenta cantidad de letras con tilde
cont_vocales = 0     #Cuenta vocales mayúsculas
cont_spaces = 0      #Cuenta espacios introducidos por el usuario en la cadena
cont_reserv = 0      #Cuenta la cantidad de palabras reservadas introducidas
cont_digits = 0      #Cuenta cantidad de dígitos en la cadena


#Se usa un ciclo for para buscar los elementos correspondientes de cada
#contador, busacando elemento por elemento en la cadena ingresada
for i in range(len(cadena_11)):
    if cadena_11[i] in vocales or cadena_11[i] in vocales_tilde:
        cont_vocales += 1 
        
    if cadena_11[i] in vocales_tilde or cadena_11[i] in vocales_tilde_minus:
        cont_tildes += 1
        
    if cadena_11[i].isdigit() is True:
        cont_digits += 1
        
    if cadena_11[i] in cad_space:
        cont_spaces += 1 

for j in lista_11:
    if j in reserv:
        cont_reserv+=1

#Se imprimen los resultados de los contadores
print('la cadena ingresada contiene: ')
print('  ',cont_vocales, 'vocales en mayúscula')
print('  ',cont_tildes, 'letras con tilde entre mayúsculas y minúsculas')  
print('  ',cont_digits, 'dígitos')     
print('  ',cont_spaces, 'espacios')   
print('  ',cont_reserv, 'palabras reservadas')
```

```python

          #Se crea una lista vacía
lista_15=[]

          #Entrada del usuario
while True:
    try:
        x_15 = int(input("Introduzca la cantidad de dígitos a ingresar: "))
        if x_15 >= 0: 
            break
        else:
           print ('La cantidad de elementos debe ser positiva.') 
    
    except ValueError:
        print ('La cantidad de elementos debe ser un valor entero')
    
#Se pide al usuario ingresar los elementos de la lista  
for i in range(x_15):
    
    while True:
        try:
            elemento = int(input("Ingrese el elemento valor entero: "))
            lista_15.append(elemento)
            break
        except ValueError:
            print ('Ingrese solo números enteros')
   
con_15 = 0
repit_15 = 0      #Cuenta el número de dígitos repetidos 2 veces

#Se usa la funcipon count() para buscar los dígitos repetidos
#Cuando count()==2 el dígito estará repetido 2 veces.
while con_15 < x_15:
    num1 = lista_15[con_15]
    contador = lista_15.count(num1)
    
    if contador == 2:
        repit_15 += 1
        con_15 += 1 
    else:    
        con_15 += 1

#Se imprime la cantidad de números repetidos 2 veces en la cadena ingresada
print ('La lista que usted ingresó es: ',lista_15)
print('En esta lista hay', int(repit_15/2), 
     'número(s) repetido(s) exactamente 2 veces.')
```

```python

#Se crea una lista vacía
lista_17=[]

              #El usuario ingresa la cantidad de elementos de la lista
while True:
    try:
        x_17 = int(input("Ingrese el numero de elementos: "))
        if x_17 > 1: 
            break
        else:
           print ('La longitud de la lista debe ser mayor a 1.') 
    
    except ValueError:
        print ('La longitud de la lista debe ser un número entero.')
    
             #El usuario digita los elementos de la lista       
for i in range(x_17):
    while True:
        try:
            elemento_17=(input("Ingrese un elemento de la lista: "))
            lista_17.append(elemento_17)
            break
        except ValueError:
            print ('Inténtelo de nuevo.')
   
          #Imprimimos la lista ingresada por el ususario         
print('La lista que usted ingresó es: ', lista_17) 

  #Con max() buscamos el elemento mayor en la lista ingresada por el usuario
mayor_lista = max(lista_17) 

#Eliminamos por completo de la lista dicho elemento mayor
while  mayor_lista in lista_17:
    lista_17.remove(mayor_lista)

#Definimos nuevamente el mayor de la lista, 
#y este será el segundo mayor de la lista inicial
mayor_lista=max(lista_17)

   #Imprimimos dicho valor
print('El segundo elemento mayor de la lista es: ', mayor_lista)
 
```
