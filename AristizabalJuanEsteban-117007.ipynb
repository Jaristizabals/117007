```python
      
suma = 0                 # Suma de los dígitos del número

def serie_Fibonacci(suma):   # Función para saber si el resultado de la suma
    a = 0                    # hace parte de la serie Fibonacci
    b = 1
    fib = 0
    while fib < suma:  #Definir serie de Fibonacci
        fib = a + b
        a = b
        b = fib
        if fib == suma:  #Cuando la suma de los dígitos 
                         #sea un número de dicha serie
            return True
    return False

x6 = int(input('Digite un número entero: '))  #Entrada del usuario

if x6 < 0:    
    x6 *= -1     #Trabajaremos con la parte positiva del número

while x6 != 0:         #Vamos realizando la suma de los dígitos del número
    suma += (x6 % 10)  #y guardamos esta suma en la variable 'suma'
    
    x6 //= 10    #Se va eliminando el último dígito del número hasta que sea 0

if serie_Fibonacci(suma):   #Cuando la suma de los dígitos sea una 
                            #número de la serie de Fibonacci
    print("La suma de sus dígitos es igual a", suma,     
           "y es un número de la serie de Fibonacci.")
else:                       #Cuando la suma de los dígitos no sea una 
                            #número de la serie de Fibonacci
    print("La suma de sus dígitos es igual a", suma,  
          "pero no es un número de la serie de Fibonacci.")

        
```
